```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
total_students = 100
```

```python
fail_in_one_subject = 20
fail_in_two_subject = 10
fail_in_three_subject = 3
```

```python
p_fail_in_one_subject = fail_in_one_subject/total_students
p_fail_in_two_subject = fail_in_two_subject/total_students
p_fail_in_three_subject = fail_in_three_subject/total_students
```

```python
x = [1,2,3]
y = [p_fail_in_one_subject,p_fail_in_two_subject,p_fail_in_three_subject]
```

```python
plt.bar(x,y)
plt.xlabel("no of subject students failed in")
plt.ylabel("probability of student failing in subjects")
```
